```python
# step-1: import libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
```

```python
# Load the imdb dataset word index

word_index = imdb.get_word_index()
revesed_word_index = {value:key for key, value in word_index.items()}
```

```python
# Load the model
model = load_model("simple_rnn_imdb.h5")
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
def preprocess_text(txt):
    words = txt.strip().lower().split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = sequence.pad_sequences([encoded_review],maxlen=500)
    return padded_review
```

```python
# prediction
def predict(review):
    encoded_review = preprocess_text(review)
    prediction = model.predict(encoded_review)
    sentiment = "Positive" if prediction[0][0] > 0.5  else "Negative"
    return sentiment, prediction[0][0]
```

```python
# test with example review

example_review = "The movie was absolutely terrible and boring"

sentiment, score = predict(example_review)

print(example_review)
print(sentiment)
print(score)
```
